```python
kickstarter_numeric = kickstarter[['state_successful','backers_count','converted_pledged_amount','goal','staff_pick','cam_length','blurb_length']]
```

```python
sns.set(font_scale=1) 
sns.pairplot(kickstarter_numeric,hue='state_successful',palette='coolwarm')
```

```python
sns.set(font_scale=1) 


sns.boxplot(x="state_successful", y="cam_length", data=kickstarter,palette='rainbow')
#sns.boxplot(x="state_successful", y="blurb_length", data=kickstarter,palette='rainbow')
#sns.countplot(x = 'staff_pick', data = kickstarter)
```

```python
sns.boxplot(x="state_successful", y="blurb_length", data=kickstarter,palette='rainbow')
```

```python
sns.boxplot(x="country", y="cam_length", hue="state_successful",data=kickstarter, palette="coolwarm")
```

```python
sns.set(font_scale=3) 
plt.figure(figsize=(45,20))
ax = sns.boxplot(x = 'main_category', y = 'cam_length', hue = 'state_successful', data=kickstarter, palette = 'rainbow')
leg = ax.get_legend()
leg.set_title("Outcome")
labs = leg.texts
labs[0].set_text("Failed")
labs[1].set_text("Succeeded")

ax.set(xlabel='Category', ylabel='Campaign Length')
#plt.savefig('eda1.svg')
#plt.savefig('eda1.png')
plt.tight_layout
```

```python
sns.set(font_scale=3) 
plt.figure(figsize=(45,20))
ax = sns.barplot(x = 'main_category', y = 'goal', hue = 'state_successful', data=kickstarter, palette = 'rainbow')
leg = ax.get_legend()
leg.set_title("Outcome")
labs = leg.texts
labs[0].set_text("Failed")
labs[1].set_text("Succeeded")

ax.set(xlabel='Category', ylabel='Goal in USD')
#plt.savefig('eda2.svg')
#plt.savefig('eda2.png')
plt.tight_layout
```
